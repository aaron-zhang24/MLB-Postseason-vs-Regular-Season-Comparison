This cell merges all the other regular season dataFrames into one, creating a single dataFrame (combined_df) containing player statistics from multiple seasons. It then calculates the average values for numeric columns for each player using the groupby method, resulting in an aggregated DataFrame (aggregated_df). Additionally, it counts the number of seasons each player contributed to (player_counts) and merges this information back into the aggregated DataFrame (result_df_rs) for a comprehensive view of player performance across seasons.

```python
all_seasons = [df_2024_reg_season, df_2023_reg_season, df_2022_reg_season, df_2021_reg_season, df_2019_reg_season, df_2018_reg_season, df_2017_reg_season, df_2016_reg_season, df_2015_reg_season, df_2014_reg_season, df_2013_reg_season, df_2012_reg_season, df_2011_reg_season, df_2010_reg_season]

# Concatenate DataFrames
regular_season_combined_df = pd.concat(all_seasons, ignore_index=True)

# Group by 'Name' and compute averages for numeric columns
aggregated_regular_season_df = regular_season_combined_df.groupby('Name').mean(numeric_only=True).reset_index()

# Display the result
aggregated_regular_season_df
```

# **Visualization Choices & Alignment With Goals**

**Scatter Plot**

This visualization highlights the differences in star players' offensive performance (OBP and SLG) between the regular season and postseason. By using a facet grid, it separates the data into two clear contexts, making it easy to compare trends side by side. The addition of ERA values in the legend provides critical context, showing how the quality of pitching (measured by ERA) differs between seasons and likely influences player performance.

The plot includes clear axis labels for OBP and SLG, consistent scaling across both facets for direct comparison, and color-coded points to distinguish between regular season and postseason data. The tighter clustering of regular season data reflects more consistent performance, while the wider spread in postseason data highlights greater variability. These elements together provide a clear and focused representation of the trends and underlying factors affecting player performance.

**Distribution**

We also wanted to visualize the distribution of ERA among pitchers to see highlight some of the summary statistics we see in the table.

**Bar Plot**

We use a bar plot visualization to first look at the on-base percentage (OBP) by player leverage level (pLI) between the regular season and postseason for star batters. We seperate the bar plot into two types of pLI levels, low and high in order to compare the differences in OBP under varying pressure situations across both seasons. By visualizing OBP in this way, we can determine whether players maintain consistent performance or exhibit significant differences when facing higher leverage scenarios in the postseason, aligning with our goal of analyzing how pressure impacts batting performance.
